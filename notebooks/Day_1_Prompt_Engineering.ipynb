## 🛠️ Smart Task 1: Summarization Prompt

**Prompt:** _"Summarize the Ramayana in 3 lines, for a 10-year-old."_

## 🛠️ Smart Task 2: Python Prompt

**Prompt:** _"Write a Python script to print even numbers from 1 to 100. Add comments."_

```python
# Python script to print even numbers from 1 to 100
for i in range(1, 101):
    if i % 2 == 0:
        print(i)  # Print even number
```
